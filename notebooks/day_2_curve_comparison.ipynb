```python
import matplotlib.pyplot as plt
from data.treasury import fetch_treasury_curve
from curves.yield_curve import YieldCurve

df = fetch_treasury_curve()

# Pick 3 dates (first, middle, last)
dates = [df.index[-3], df.index[-2], df.index[-1]]

fig, ax = plt.subplots()
for d in dates:
    yc = YieldCurve(d, df.loc[d])
    yc.plot(ax=ax)

ax.legend([str(d.date()) for d in dates])
plt.show()
```
